```python
import pandas as pd
import matplotlib.pyplot as plt

# read file and group
filename = "./visual/data.csv"
x_labels = ["set 1", "set 2", "set 3"]
df = pd.read_csv(filename)
grouped_df = df.groupby(["set", "model"])["loss"].mean().reset_index()
print(grouped_df)
```

```python
# draw barplot
plt.figure(figsize=(10, 6))
bar_width = 0.20

# Set positions for bars on X-axis
bar_positions = range(1, len(grouped_df["set"].unique()) + 1)

# Plot bars for each model
for i, model in enumerate(grouped_df["model"].unique()):
    model_data = grouped_df[grouped_df["model"] == model]
    plt.bar([p + i * bar_width for p in bar_positions], model_data["loss"], bar_width, label=model)

# plt.xlabel("Set")
plt.ylabel("Mean Squared Error")
# plt.title("Average Loss by Set and Model")
# plt.xticks(bar_positions, ["Set {}".format(s) for s in grouped_df["set"].unique()])
plt.legend()
plt.tight_layout()
plt.show()
```
