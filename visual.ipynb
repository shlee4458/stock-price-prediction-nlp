```python
import pandas as pd
import matplotlib.pyplot as plt

# read file and group
filename = "./visual/data.csv"
x_labels = ["set 1", "set 2", "set 3"]
df = pd.read_csv(filename)
grouped_df = df.groupby(["set", "model"])["loss"].mean().reset_index()

print(grouped_df)
```

```python
# Define colors
colors = {"LSTM": "#2D8FCF", "XGBoost": "orange"}

# Set figure background to black
fig, ax = plt.subplots(figsize=(10, 6), facecolor='black')

# Set bar width and positions
bar_width = 0.20
bar_positions = range(1, len(grouped_df["set"].unique()) + 1)

# Plot bars for each model
for i, model in enumerate(grouped_df["model"].unique()):
    model_data = grouped_df[grouped_df["model"] == model]
    ax.bar([p + i * bar_width for p in bar_positions], model_data["loss"], bar_width, color=colors[model], label=model)

# Customize labels and ticks
ax.set_ylabel("Mean Squared Error", color='white')
ax.tick_params(axis='y', colors='white')
ax.tick_params(axis='x', colors='white')
ax.set_xticks([])

# Add legend
ax.legend(facecolor='black', loc='upper right', bbox_to_anchor=(1, 1), fontsize='large', labelcolor='white')

# Remove grid
ax.grid(False)

# Set facecolor of the axes
ax.set_facecolor('black')

# Show plot
plt.tight_layout()
plt.show()
```
